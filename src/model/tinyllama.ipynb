```python
from transformers import AutoProcessor, AutoModelForImageTextToText

processor = AutoProcessor.from_pretrained("meta-llama/Llama-4-Scout-17B-16E-Instruct")
model = AutoModelForImageTextToText.from_pretrained("meta-llama/Llama-4-Scout-17B-16E-Instruct")
messages = [
    {
        "role": "user",
        "content": [
            {"type": "image", "url": "https://huggingface.co/datasets/huggingface/documentation-images/resolve/main/p-blog/candy.JPG"},
            {"type": "text", "text": "What animal is on the candy?"}
        ]
    },
]
inputs = processor.apply_chat_template(
	messages,
	add_generation_prompt=True,
	tokenize=True,
	return_dict=True,
	return_tensors="pt",
).to(model.device)

outputs = model.generate(**inputs, max_new_tokens=40)
print(processor.decode(outputs[0][inputs["input_ids"].shape[-1]:]))
```

## TinyLlama com verificador

```python
from transformers import pipeline
import json, re

# ----------------------------------------
# 1) Assistente EXTRATOR
# ----------------------------------------
extractor = pipeline("text-generation", model="TinyLlama/TinyLlama-1.1B-Chat-v1.0")

def apply_chat_template(pipe, messages):
    return pipe.tokenizer.apply_chat_template(messages, tokenize=False, add_generation_prompt=True)

def only_json_block(text: str) -> str:
    m = re.search(r"\{.*\}", text, flags=re.DOTALL)
    return m.group(0) if m else text.strip()

def run_extractor(email_body: str) -> str:
    messages = [
        {
            "role": "system",
            "content": """
Você é um assistente de extração de dados. Sua tarefa é analisar o corpo de uma mensagem de e-mail e extrair três informações-chave: o nome do ativo financeiro, o preço unitário e a quantidade.

A resposta deve ser **estritamente** um dicionário JSON, seguindo este formato:
{"Nome": string, "Preco": float, "Qtd": integer}. A resposta deve ser apenas o JSON, sem nenhum comentário acicional

Caso alguma informação não seja encontrada, o valor deve ser `null`.
Caso o preço esteja em reais (R$), remova o símbolo.

---
Exemplos:

Exemplo 1 (entrada → saída):
"Olá, a equipe. A ordem de compra para 100 ações da Petrobras (PETR4) foi executada. O preço por ação foi de 32.50."
{"Nome": "Petrobras (PETR4)", "Preco": 32.50, "Qtd": 100}

Exemplo 2 (entrada → saída):
"Compra de 500 cotas do IVVB11 a 235,45."
{"Nome": "IVVB11", "Preco": 235.45, "Qtd": 500}

Agora, analise o texto a seguir:
""".strip(),
        },
        {
            "role": "user",
            "content": email_body.strip(),
        },
    ]

    prompt = apply_chat_template(extractor, messages)
    out = extractor(prompt, max_new_tokens=300, do_sample=True, temperature=0.4, top_k=40, top_p=0.9)[0]["generated_text"]
    # remove o prompt ecoado (alguns modelos retornam prompt + geração)
    if out.startswith(prompt):
        out = out[len(prompt):]
    return only_json_block(out).strip()

# ----------------------------------------
# 2) Assistente VERIFICADOR (LLM)
#    - checa se o JSON do extrator cumpre o contrato
#    - responde ESTRITAMENTE: {"valido": bool, "motivo": string|null}
# ----------------------------------------
verifier = pipeline("text-generation", model="TinyLlama/TinyLlama-1.1B-Chat-v1.0")

VER_RULES = """
Contrato de saída esperado do EXTRATOR:
- Deve ser ESTRITAMENTE um objeto JSON (sem texto extra) com chaves exatas: "Nome", "Preco", "Qtd".
- Tipos: "Nome" = string ou null; "Preco" = número (float) ou null; "Qtd" = número inteiro ou null.
- O preço não deve conter "R$" (apenas número, ponto como separador decimal).
- Se algum campo não for encontrado, deve ser null (JSON null).
- Não pode haver comentários, explicações, sufixos ou prefixos fora do JSON.
""".strip()

def run_verifier(email_body: str, extrator_json_text: str) -> str:
    messages = [
        {
            "role": "system",
            "content": f"""
Você é um verificador de conformidade de saída JSON.
Tarefa: Dizer se o JSON produzido pelo assistente EXTRATOR está de acordo com o contrato.

Sua resposta deve ser ESTRITAMENTE um JSON no formato:
{{"valido": (true ou false)}}


{VER_RULES}
""".strip(),
        },
        {
            "role": "user",
            "content": f"""
TEXTO_ANALISADO:
{email_body}

SAIDA_DO_EXTRATOR (texto bruto):
{extrator_json_text}
""".strip(),
        },
    ]
    prompt = apply_chat_template(verifier, messages)
    out = verifier(prompt, max_new_tokens=200, do_sample=False, temperature=0.0)[0]["generated_text"]
    if out.startswith(prompt):
        out = out[len(prompt):]
    return only_json_block(out).strip()

# ----------------------------------------

if __name__ == "__main__":
    email = "COMPRA:NTN-B 15/08/2026QTDE: 320.000TAXA: 18,2990%PU: 4.374,142915FIN: R$ 999.999,45LIQ: 21/08/2025"

    extracao = run_extractor(email)
    print("Extrator →", extracao)

    validacao = run_verifier(email, extracao)
    print("Verificador →", validacao)

```

```python
email = "COMPRA:NTN-B 15/08/2026QTDE: 350.000TAXA: 18,2990%PU: 4.374,142915FIN: R$ 999.999,45LIQ: 21/08/2025"

extracao = run_extractor(email)
print("Extrator →", extracao)

validacao = run_verifier(email, extracao)
print("Verificador →", validacao)
```
